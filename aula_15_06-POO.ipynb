```python
class produto():
  descricao = str
  qtd = int
  
  def __init__(self, descricao=None, qtd=None):
    if descricao is None:
      self.descricao = ''
    else:
      self.descricao = descricao

    if qtd is None:
      self.qtd = 0

    else:
      self.qtd = qtd

  def leituradados(self):
    self.descricao = str(input('Digite o nome do protudo: '))
    self.qtd = int(input('Digite a quantidade: '))

  def mostradados(self):
    print('Produto:{}, quantidade:{} '.format(self.descricao, self.qtd))

  def saida(self):
    saiu = int(input('Digite a saida total: '))
    if(self.qtd<=0):
      print('Impossivel retirar!')
      return
    self.qtd -=saiu;
  
  def entrada(self):
    entrou= int (input('Digite o valor que entrou: '))
    self.qtd += entrou
```

```python
produto1 = produto('sabao', 2)
```

```python
produto1.mostradados()
```

```python
produto1.saida()
```
