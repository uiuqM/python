Faça um programa que leia três números inteiros. Após, determine e escreva na tela qual o menor,
o do meio e o maior.

```python
n= float(input())
n1= float(input())
n2= float(input())

if(n>n1 and n>n2):
  if(n1>n2):
    print("maior:{} meio: {} menor:{}".format(n, n1, n2))
  else:
    print("maior:{} meio: {} menor:{}".format(n, n2, n1))

elif(n1>n and n1>n2):
  if(n>n2):
    print("maior:{} meio: {} menor:{}".format(n1, n, n2))
  else:
    print("maior:{} meio: {} menor:{}".format(n1, n2, n))
else:
  if(n1>n):
    print("maior:{} meio: {} menor:{}".format(n2, n1, n))
  else:
    print("maior:{} meio: {} menor:{}".format(n2, n, n1))

#Usando o método sort p listar do maior p menor
lista = [n, n1, n2]
lista.sort(reverse=True)

print("maior={} meio={} menor={}".format(lista[0], lista[1], lista[2]))
```

Faça um programa que leia um número qualquer. Após divida-o por 10 várias vezes, até que o
resultado da divisão seja considerado 0 pelo computador. No processo, conte quantas vezes essa
divisão foi realizada e apresente esse valor na tela.

```python
n=int(input())
cont = 0
div = True
while(div):
  n = n/10
  cont+=1
  if(n==0):
    div = False
  
print(cont)
```

Faça um programa que leia um número inteiro positivo e que calcule o valor do fatorial desse
número, fazendo uso de uma função. O programa, só deve aceitar a leitura de um número inteiro
positivo, ou seja, enquanto não for informado um número inteiro positivo, deve-se continuar a
leitura do mesmo.

```python
def fatorial(num):
  seq = num
  for i in range(1, num-1):
    seq*=i
  return seq
n = int(input())
fatorial(n)
```
